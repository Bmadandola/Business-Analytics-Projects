# This is Decision Tree for Data Mining

```python
import pandas as pd #This is pandas library
```

```python
df = pd.read_csv('C:\\Users\\biola\\Downloads\\decisionTree_Data.csv')
```

```python
X = df[['Age', 'Salary']]  # Age and Salary are the two predictors 
```

```python
df.head()

```

```python
df.describe()
```

```python
y = df['Purchased']  # Target Variable 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion='gini', random_state=42)  # Create and train the model

clf.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=X.columns, class_names=['Purchased', 'Not Purchased'], filled=True)  # Modify class names
plt.show()
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
